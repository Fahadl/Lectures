```python
order_of_bars
```

```python
# Tornado Plot. 

for c, group in zip(colors, tips[group_col].unique()):
    sns.barplot(x='size', y='day', data=tips.loc[tips[group_col]==group, :],
              order=order_of_bars, color=c,
              label=group, errwidth=0)
    plt.legend()
```

## Outliers Handling

```python
sns.boxplot(data=tips);
```

```python
Q1 = np.percentile(tips['tip'], 25, interpolation = 'midpoint')
Q3 = np.percentile(tips['tip'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
               
print('lower bound outliers ', Q1 - 1.5*(IQR))
print('higher bound outliers', Q3 + 1.5*(IQR))
```

```python
tips2 = tips.drop(tips[(tips['tip'] > 5.9375) | (tips['tip'] < -0.36250)].index)
```

```python
sns.boxplot(data=tips2);
```
